```python
pd.read_sql_query("""
SELECT
    payment_method,
    COUNT(*) AS number_of_transactions,
    SUM(quantity) AS total_quantity_sold,
    (COUNT(*) - SUM(quantity)) AS difference
FROM sales
GROUP BY payment_method
ORDER BY payment_method
""", conn)
```

```python
pd.read_sql_query("""
WITH avg_rating AS (
    SELECT
        branch,
        category,
        AVG(rating) AS avg_rating
    FROM sales
    GROUP BY branch, category
),
ranked AS (
    SELECT
        branch,
        category,
        avg_rating,
        RANK() OVER (PARTITION BY branch ORDER BY avg_rating DESC) AS rnk
    FROM avg_rating
)
SELECT
    branch,
    category,
    ROUND(avg_rating, 2) AS avg_rating
FROM ranked
WHERE rnk = 1
""", conn)
```

```python
pd.read_sql_query("""
WITH daily_transactions AS (
    SELECT
        branch,
        -- Convert DD/MM/YY to YYYY-MM-DD format for strftime, assuming consistent 2-digit day/month/year parts
        strftime('%A', '20' || SUBSTR(TRIM(date), 7, 2) || '-' || SUBSTR(TRIM(date), 4, 2) || '-' || SUBSTR(TRIM(date), 1, 2)) AS day_name,
        COUNT(*) AS total_transactions
    FROM sales
    -- Filter out rows where date format is not 'DD/MM/YY' (length 8) to avoid parsing errors
    WHERE LENGTH(TRIM(date)) = 8
    GROUP BY branch, day_name
),
ranked_days AS (
    SELECT
        branch,
        day_name,
        total_transactions,
        RANK() OVER (PARTITION BY branch ORDER BY total_transactions DESC) AS rnk
    FROM daily_transactions
    WHERE day_name IS NOT NULL -- Keep this filter as a safety net for any remaining invalid date conversions
)
SELECT
    branch,
    day_name AS busiest_day,
    total_transactions
FROM ranked_days
WHERE rnk = 1
""", conn)
```

```python
pd.read_sql_query("""
SELECT
    payment_method,
    SUM(quantity) AS total_quantity_sold
FROM sales
GROUP BY payment_method
ORDER BY total_quantity_sold DESC
""", conn)
```

```python
pd.read_sql_query("""
SELECT
    city,
    ROUND(AVG(rating), 2) AS avg_rating,
    MIN(rating) AS min_rating,
    MAX(rating) AS max_rating
FROM sales
GROUP BY city
ORDER BY avg_rating DESC
""", conn)
```

```python
pd.read_sql_query("""
SELECT
    category,
    ROUND(SUM(unit_price * quantity * profit_margin), 2) AS total_profit
FROM sales
GROUP BY category
ORDER BY total_profit DESC
""", conn)
```

```python
pd.read_sql_query("""
WITH payment_count AS (
    SELECT
        branch,
        payment_method,
        COUNT(*) AS payment_count
    FROM sales
    GROUP BY branch, payment_method
),
ranked_payment AS (
    SELECT
        branch,
        payment_method,
        payment_count,
        RANK() OVER (PARTITION BY branch ORDER BY payment_count DESC) AS rnk
    FROM payment_count
)
SELECT
    branch,
    payment_method AS preferred_payment_method
FROM ranked_payment
WHERE rnk = 1
""", conn)
```

```python
pd.read_sql_query("""
SELECT
    CASE
        WHEN time BETWEEN '00:00:00' AND '11:59:59' THEN 'Morning'
        WHEN time BETWEEN '12:00:00' AND '17:59:59' THEN 'Afternoon'
        ELSE 'Evening'
    END AS shift,
    COUNT(invoice_id) AS number_of_invoices
FROM sales
GROUP BY shift
ORDER BY number_of_invoices DESC
""", conn)
```

```python
pd.read_sql_query("""
WITH yearly_revenue AS (
    SELECT
        branch,
        strftime('%Y', date) AS year,
        SUM(quantity * CAST(REPLACE(unit_price, '$', '') AS REAL)) AS revenue
    FROM sales
    WHERE strftime('%Y', date) IN ('2022', '2023')
    GROUP BY branch, year
),
pivoted AS (
    SELECT
        branch,
        SUM(CASE WHEN year = '2022' THEN revenue ELSE 0 END) AS revenue_2022,
        SUM(CASE WHEN year = '2023' THEN revenue ELSE 0 END) AS revenue_2023
    FROM yearly_revenue
    GROUP BY branch
),
decrease_calc AS (
    SELECT
        branch,
        revenue_2022,
        revenue_2023,
        ROUND(
            (revenue_2022 - revenue_2023) * 1.0 / revenue_2022,
            4
        ) AS decrease_ratio
    FROM pivoted
    WHERE revenue_2023 < revenue_2022
)
SELECT
    branch,
    revenue_2022,
    revenue_2023,
    decrease_ratio
FROM decrease_calc
ORDER BY decrease_ratio DESC
LIMIT 5
""", conn)
```
